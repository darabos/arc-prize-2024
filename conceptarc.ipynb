```python
import json
with open('ConceptARC/corpus/AboveBelow/AboveBelow7.json') as f:
  data = json.load(f)


from IPython.display import display, HTML
colors = ['white', 'black', 'blue', 'red', 'green', 'yellow', 'purple', 'orange', 'brown', 'pink', 'gray', 'cyan']
def show_board(board):
  html = '<table>'
  for row in board:
    html += '<tr>'
    for cell in row:
      color = colors[cell]
      html += f'<td style="padding: 0; width: 10px; height: 10px; background-color: {color};"></td>'
    html += '</tr>'
  html += '</table>'
  return html

def show_example(example):
  html = '<table><tr><td>'
  html += show_board(example['input'])
  html += '</td><td>→</td><td>'
  if 'output' in example:
    html += show_board(example['output'])
  html += '</td></tr></table>'
  return html

def show_task(task):
  if isinstance(task, int):
    task = data[list(data.keys())[task]]
  for t in task:
    h = show_example(t)
    display(HTML(h))

print(len(data.keys()))
show_task(0)
```
